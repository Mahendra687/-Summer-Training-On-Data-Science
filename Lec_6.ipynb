#### Scatter plot___
to compare the variables

#### 3D Plot

```python
from pandas import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
iris = read_csv("IRIS.csv")
iris.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = "3d")

for species, irisset in iris.groupby("species"):
    ax.scatter(irisset['petal_length'],irisset['sepal_length'],irisset['petal_width'],label = species)
    ax.set_xlabel('P_lenth')
    ax.set_ylabel('sepal_length')
    ax.set_zlabel("P_width")
    
    plt.show()
```
