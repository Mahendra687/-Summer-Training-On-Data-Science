```python
from pandas import*
from numpy import*
import pandas as pd
import numpy as np
from sklearn import*
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({'x1':[12,20,28,18,29,33,24,45,43,51,52,55,53,54,61,64,69,52,6,62,45,60,59,8], 'x2':[39,36,30,52,54,46,55,59,63,70,66,63,58,14,8,19,7,24,77,65,80,4,72,45]})
```

```python
from sklearn.cluster import KMeans
```

```python
data.head()
```

```python
#import KM algorithm
```

```python
km = KMeans(n_clusters = 3)#cluster of three
```

```python
# Fit our data in this algorithm
```

```python
X = data
X.head()
```

```python
km.fit(X)
```

```python
# Check the Centroid values
```

```python
km.cluster_centers_ # number of cluster centre is 3
```

```python
# Check out the Labels 
```

```python
km.labels_ # labels for all the data
```

```python
data['K means predicted labels'] = km.labels_
data.head()
```

```python
labels = km.labels_
labels
```

```python
plt.plot(data['x1'],data['x2'])
plt.show()
```

```python
plt.scatter(data['x1'],data['x2'])
plt.show()
```

```python
# if we have many features like x1,x2.....Xn in this case we can select the any of two features
```

```python
# assign the color for each cluster 
```

```python
plt.scatter(data['x1'],data['x2'],40,labels)
plt.show() # 40 is the size and for the colors we put the labels..
```

```python
# ploting the centoid points
plt.scatter(data['x1'],data['x2'],40,labels)


for i in km.cluster_centers_:
    plt.scatter(i[0],i[1],100) #100 is the marker size 
    plt.show()
```

```python
# Elbow Method...
```

```python
wss = []
cluster_range = range(1,10)

for k in cluster_range:
    km = KMeans(n_clusters=k)
    km.fit(X)
    inertia = km.inertia_ #inertia is the closest centre
    wss.append(inertia)
    
    print(wss)
```

```python
plt.xlabel('k, here number of cluster')
plt.ylabel('Sum of Squre Error')

plt.plot(cluster_range,wss)
plt.show()
```
